```python
import pandas as pd

df = pd.read_csv("student-por.csv")

df.head()
avg_score=df['G3'].mean()
avg_score

above_avg_student=df[df['G3']>avg_score]
above_avg_student

below_avg_student=df[df['G3']<avg_score]
print("Above Average Students:")
print(above_avg_student)
print("\nBelow Average Students:")
print(below_avg_student)
```

```python
#top 5 student
top_5_student=df.nlargest(5,'G3')
print("Top 5 students:")
print(top_5_student)
```

```python
students=df.sort_values(by='G3',ascending=False)
print(students)
```

```python
# top 5 students in descending
students=df.sort_values(by='G3',ascending=False).head(5)[['G1','G2','G3']]
print(students)
```

```python
# find the how many student passed the final exam
passing_score = 10
passed_students = df[df['G3'] >= passing_score]
num_passed_students = len(passed_students)

print(f"Number of students who passed the final exam (G3 >= {passing_score}): {num_passed_students}")
```

```python
#find the sudent with 0 absences
df[df['absences'] == 0]
```

```python
#find the students who improved for G1 to G3
df[df['G3'] > df['G1']]
```

```python
# find the student consistent performance
df[(df['G1'] == df['G2']) & (df['G2'] == df['G3'])][['G1','G2','G3']]
```

```python
#matplot lib
#matpot lib tool is helpful for visulaisation
import matplotlib.pyplot as plt
plt.figure()
plt.hist(df['G3'],bins=10)
```

```python
#matplot lib
#matpot lib tool is helpful for visulaisation
# histogram
import matplotlib.pyplot as plt
plt.figure()
plt.hist(df['G3'],bins=10)
plt.xlabel("G3")
plt.ylabel("Frequency")
plt.title("Histogram of G3")
plt.show()
```

```python
# bar chart
# avg first mark scored by boys and grils
avg_scored_by_gender=df.groupby('sex')['G1'].mean()
plt.figure()
avg_scored_by_gender.plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Average First Mark")
plt.title("Average First Mark by Gender")
plt.show()
```

```python
#scatter plot
plt.figure()
plt.scatter(df['studytime'],df['G3'])
plt.xlabel("studytime")
plt.ylabel("G3")
plt.title("Scatter Plot of studytime vs G3")
plt.show()
```

```python
#box plot
plt.figure()
plt.boxplot(df['G3'])
plt.ylabel("G3")
plt.title("Box Plot of G3")
plt.show()
```

```python
# line plot
avg_marks=df[['G1','G2','G3']].mean()
plt.figure()
plt.plot(avg_marks.index,avg_marks.values,marker='o')
plt.xlabel("subject")
plt.ylabel("average marks")
plt.title("average marks by subject")
plt.show()
```

```python
#with bar chart we will find the number of student
# who failed and passed the final exam
pass_ = df['G3']>=10
plt.figure()
plt.bar(['pass','fail'],[pass_.sum(),len(df)-pass_.sum()])
plt.xlabel("result")
plt.ylabel("number of students")
plt.title("Number of students who passed and failed the final exam")
plt.show()
```

```python
#histogram absence distribution
plt.figure()
plt.hist(df["absences"],bins=10)
plt.xlabel("absences")
plt.ylabel("frequency")
plt.title("Histogram of absences")
plt.show()
```
